```python
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client['atividade']
collection = db['restaurante']

# Consulta para encontrar restaurantes no Bronx que preparem cozinha americana ou chinesa
bronx_americana_chinesa = collection.find({
    "borough": "Bronx",
    "$or": [{"cuisine": "American "}, {"cuisine": "Chinese"}]
})

for restaurante in bronx_americana_chinesa:
    print("restaurant_id:", restaurante["restaurant_id"])
    print("name:", restaurante["name"])
    print("borough:", restaurante["borough"])
    print("cuisine:", restaurante["cuisine"])
    print("\n")

client.close()
```

```python
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client['atividade']
collection = db['restaurante']

pipeline = [
    {
        "$group": {
            "_id": "$cuisine",
            "boroughs": {"$addToSet": "$borough"}
        }
    },
    {
        "$unwind": "$boroughs"
    },
    {
        "$sort": {
            "_id": 1,
            "boroughs": -1
        }
    },
    {
        "$project": {
            "cuisine": "$_id",
            "borough": "$boroughs",
            "_id": 0
        }
    }
]

restaurantes_ordenados = list(collection.aggregate(pipeline))

for restaurante in restaurantes_ordenados:
    print("cuisine:", restaurante["cuisine"])
    print("borough:", restaurante["borough"])
    print("\n")

client.close()
```
